```python
import gspread
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
service_account = gspread.service_account(filename="/Users/timal/.config/gspread/service_account.json")
xom_chpt_sheet = service_account.open('XOM_CHPT')
work_sheet = xom_chpt_sheet.worksheet('XOM_CHPT')
```

```python
df = pd.DataFrame(work_sheet.get_all_values())
```

```python
df.rename(columns=df.iloc[0], inplace=True)
df.drop(df.index[0], inplace=True)
df
```

```python
%matplotlib inline
plt.figure(figsize=(80,80))
plt.scatter(df.XOM, df.CHPT)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.XOM, df.CHPT, test_size=0.2, random_state=10)
clf = LinearRegression()
clf.fit(x_train.values.reshape(-1,1), y_train)
```

```python
clf.score(x_test.values.reshape(-1,1), y_test)
```
