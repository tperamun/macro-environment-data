```python
import pandas as pd

data = pd.read_pickle('FED_dtm.pkl')
data.head()
```

```python
data = data.transpose()
data.head()
```

```python
# find top 30 words said by each account

top_dict = {}

for u in data.columns:
    
    top = data[u].sort_values(ascending=False).head(30)
    top_dict[u] = list(zip(top.index, top.values))

top_dict
```
